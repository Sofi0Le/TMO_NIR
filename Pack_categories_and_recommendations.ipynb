### Пример рекомендательной системы

```python
import numpy as np
from sklearn.metrics.pairwise import cosine_similarity

# Функция для рекомендации
def recommend_packs(vec_list, played_pack=None):
    # Если пользователь не играл, то возвращаем первые два пака из списка
    if played_pack is None:
        return [1, 2]
    
    # Проверяем, играл ли пользователь в указанный пак
    if played_pack >= 1 and played_pack <= 5:
        played_idx = played_pack - 1
        played_vec = vec_list[played_idx]
        # Вычисляем сходство с каждым паком и сортируем по убыванию
        similarities = [cosine_similarity([played_vec], [vec])[0][0] for vec in vec_list]
        sorted_indices = np.argsort(similarities)[::-1]
        # Исключаем пак пользователя из списка рекомендаций
        sorted_indices = [idx for idx in sorted_indices if idx != played_idx]
        #print(sorted_indices)
        sorted_indices_new = [idx + 1 for idx in sorted_indices]
        #print(sorted_indices_new)
        # Возвращаем два наиболее похожих пака
        return sorted_indices_new[:2]
    else:
        print("Ошибка: номер пака должен быть от 1 до 5.")
        return []

# Пример использования
pack_num = 2
recommendations = recommend_packs(vec_list, played_pack=pack_num)
print(f"Рекомендации для пользователя, игравшего в пак {pack_num}:", recommendations)
```
